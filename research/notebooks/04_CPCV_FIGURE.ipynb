# Combinatorial Purged Cross-Validation (CPCV)

Leakage-free cross-validation for time-series / event-based labels — the validation
backbone p6lab uses. Each fold **purges** train observations whose label window overlaps
the test set, and **embargoes** a gap immediately after it (López de Prado, AFML §7).
